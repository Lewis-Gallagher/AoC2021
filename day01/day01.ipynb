```python
test = [199,200,208,210,200,207,240,269,260,263]

up = 0
down = 0

for i, j in more_itertools.pairwise(test):
    if j > i:
        up += 1
        print(i,j,"up")
    else:
        down += 1
        print(i,j,"down")

print(up)
```

```python
up = 0
down = 0

with open("day01/input.txt", "r") as depth:

    depth = depth.read().splitlines()
    
    for i, j in more_itertools.pairwise(depth):
        if j > i:
            up += 1
            print(i,j,"\tup")
        else:
            down += 1
            print(i,j,"\tdown")

print(up)
print(down)
print(up+down)
```

```python
up = 0
down = 0

with open("day01/input.txt", "r") as depth:

    depth = depth.read().splitlines()
    
    for i, j in more_itertools.pairwise(depth):
        if int(j) > int(i):
            up += 1
            print(i,j,"\tup")
        else:
            down += 1
            print(i,j,"\tdown")

print(up)
print(down)
print(up+down)
```
